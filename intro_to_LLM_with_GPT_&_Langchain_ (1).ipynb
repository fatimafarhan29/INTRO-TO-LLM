```python
import pandas as pd

def analyze_electric_car_combinations(electric_cars: pd.DataFrame) -> pd.DataFrame:
    """
    Analyzes the electric car registration data to find the counts of each make/model combination.

    Args:
        electric_cars (pd.DataFrame): A pandas DataFrame containing electric car registration data
                                       with columns 'make', 'model', and 'n_cars'.

    Returns:
        pd.DataFrame: A DataFrame containing the counts of each make/model combination,
                      sorted in descending order by count.  The DataFrame has columns 'make', 'model', and 'n_cars'.
    """

    # Group by 'make' and 'model' and sum the 'n_cars' for each combination
    make_model_counts = electric_cars.groupby(['make', 'model'])['n_cars'].sum().reset_index()

    # Rename the 'n_cars' column to 'count' for clarity
    #make_model_counts = make_model_counts.rename(columns={'n_cars': 'count'}) # No need to rename, as the prompt asks for 'n_cars'

    # Sort the results in descending order based on the count
    make_model_counts = make_model_counts.sort_values(by='n_cars', ascending=False)

    return make_model_counts


if __name__ == '__main__':# Create a sample DataFrame (replace with your actual data loading)
    data = {
        'city': ['Seattle', 'Seattle', 'Bellevue', 'Redmond', 'Seattle', 'Bellevue'],
        'county': ['King', 'King', 'King', 'King', 'King', 'King'],
        'model_year': [2020, 2020, 2020, 2020, 2019, 2020],
        'make': ['TESLA', 'NISSAN', 'TESLA', 'TESLA', 'NISSAN', 'BMW'],
        'model': ['MODEL 3', 'LEAF', 'MODEL 3', 'MODEL Y', 'LEAF', 'I3'],
        'electric_vehicle_type': ['BEV', 'BEV', 'BEV', 'BEV', 'BEV', 'BEV'],
        'n_cars': [100, 50, 75, 60, 40, 30]
    }
    electric_cars_df = pd.DataFrame(data)

    # Analyze the data
    make_model_counts_df = analyze_electric_car_combinations(electric_cars_df)

    # Print the results
    print(make_model_counts_df)
```

```python
import pandas as pd

# Assuming 'electric_cars' DataFrame is already loaded

def find_make_model_combinations(df):
    """
    Finds and returns a pandas Series containing the unique make/model combinations
    of electric cars in the given DataFrame, along with the total number of each
    combination.

    Args:
        df (pd.DataFrame): The DataFrame containing electric car registration data
                           with 'make', 'model', and 'n_cars' columns.

    Returns:
        pd.Series: A pandas Series where the index is the "make - model" combination
                   and the values are the total number of cars for that combination.
                   The Series is sorted in descending order by the number of cars.
    """

    # Combine 'make' and 'model' columns to create a new column
    # This column will represent unique make/model combinations.
    df['make_model'] = df['make'] + ' - ' + df['model']

    # Group by 'make_model' and sum the 'n_cars' for each group
    make_model_counts = df.groupby('make_model')['n_cars'].sum()

    # Sort the results in descending order to show the most common combinations first
    make_model_counts_sorted = make_model_counts.sort_values(ascending=False)

    return make_model_counts_sorted



# Example Usage:
if __name__ == '__main__':
    # Create a sample DataFrame (replace with your actual data loading)
    data = {
        'city': ['Seattle', 'Seattle', 'Bellevue', 'Bellevue', 'Redmond', 'Redmond'],
        'county': ['King', 'King', 'King', 'King', 'King', 'King'],
        'model_year': [2020, 2019, 2020, 2018, 2020, 2019],
        'make': ['Tesla', 'Tesla', 'Nissan', 'Nissan', 'Tesla', 'Nissan'],
        'model': ['Model 3', 'Model 3', 'Leaf', 'Leaf', 'Model Y', 'Leaf'],
        'electric_vehicle_type': ['BEV', 'BEV', 'BEV', 'BEV', 'BEV', 'BEV'],
        'n_cars': [100, 80, 50, 40, 70, 60]
    }
    electric_cars = pd.DataFrame(data)

    # Find the make/model combinations
    make_model_counts = find_make_model_combinations(electric_cars)

    # Print the results
    print("Make/Model Combinations and Counts:")
    print(make_model_counts)
```

```python
import pandas as pd
import plotly.express as px

def plot_ev_counts_by_year(df):
    """
    Generates a bar plot showing the total count of electric cars by model year,
    with bars colored by electric vehicle type (BEV or PHEV) using Plotly Express.

    Args:
        df (pd.DataFrame): The DataFrame containing electric car registration data
                           with 'model_year', 'electric_vehicle_type', and 'n_cars' columns.

    Returns:
        plotly.graph_objects.Figure: A Plotly Figure object representing the bar plot.
                                      The figure is also displayed using fig.show().
    """

    # Group the data by 'model_year' and 'electric_vehicle_type' and sum the 'n_cars'
    grouped_data = df.groupby(['model_year', 'electric_vehicle_type'])['n_cars'].sum().reset_index()

    # Create the bar plot using Plotly Express
    fig = px.bar(
        grouped_data,
        x='model_year',
        y='n_cars',
        color='electric_vehicle_type',
        barmode='group',  # Group bars for each year side-by-side
        title='Electric Car Counts by Model Year and Vehicle Type',
        labels={'n_cars': 'Number of Cars', 'model_year': 'Model Year', 'electric_vehicle_type': 'Vehicle Type'}
    )

    # Customize the layout (optional)
    fig.update_layout(
        xaxis_title='Model Year',
        yaxis_title='Number of Cars',
        legend_title='Vehicle Type',
        template="plotly_white"  # Optional: Use a clean white background
    )

    # Display the plot
    fig.show()

    return fig


if __name__ == '__main__':
    # Create a sample DataFrame (replace with your actual data loading)
    data = {
        'city': ['Seattle', 'Seattle', 'Bellevue', 'Bellevue', 'Redmond', 'Redmond', 'Seattle', 'Bellevue'],
        'county': ['King', 'King', 'King', 'King', 'King', 'King', 'King', 'King'],
        'model_year': [2020, 2019, 2020, 2018, 2020, 2019, 2020, 2018],
        'make': ['Tesla', 'Tesla', 'Nissan', 'Nissan', 'Tesla', 'Nissan', 'BMW', 'Ford'],
        'model': ['Model 3', 'Model 3', 'Leaf', 'Leaf', 'Model Y', 'Leaf', 'i3', 'Fusion'],
        'electric_vehicle_type': ['BEV', 'BEV', 'BEV', 'BEV', 'BEV', 'BEV', 'BEV', 'PHEV'],
        'n_cars': [100, 80, 50, 40, 70, 60, 30, 20]
    }
    electric_cars = pd.DataFrame(data)

    # Generate and display the plot
    plot_ev_counts_by_year(electric_cars)
```
